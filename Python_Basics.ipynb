```python
String Interpolation - Insert a variable into a string


- name = ‘Billy’
- print(‘Hello, my name is {}’.format(name))
- number = 27
- print(‘Hello, my name is {x} and my number is {y}’.format(x=name, y=number))

Lists - Pythons version of an array


- my_list = [‘hi’, 0, 1]
- my_list.append(’2')
- my_list.pop() < remove last item, add item location to parenthesis to remove indexed item
- 1 in my_list  - returns true or false
- Indexing in python (supports negative indexing as well eg. [-1] to get the end of the array)  - my_list[0] returns ‘hi’
- Slicing a list, add up to but not including last position - my_list[0:2] returns [‘hi’, 0]
- Nested list - nested = [1, 2, [’a’, ‘b’]] - nested[2] - returns [’a’, ‘b’] - nested[2][0] - returns ‘a’

Dictionaries - stores as key value pairs, not ordered


- dict = {‘key’: ‘name’, ‘key2’: 10}
- dict[‘key2’] - returns 10

Booleans with capitals - True, False 
Tuples - Same as arrays but they’re immutable and does not support reassignment
Set - Unordered collection of unique items - set([1,1,1,1,12,2,2,2,3,3,3,]) - returns - {1,2,3}

Comparison operators - control flow


- if 1==2:
        print(‘hello’)
    elif 2==2
        print(‘that is correct’)
    else:
        print(‘not true’)

For Loops 

- seq = [1, 2, 3, 4, 5]
- for whatever_variable_name in seq:
        print(item**2)

While Loops

- i = 1
    while i < 5 :
        print(‘i is currently {}’.format(i))
        i = i + 1

i is currently 1 > 2,3 etc 

Ranges

- range(5) returns - range(0, 5)
- for item in range(5)
    - print(item)

0,1,2,3,4

- list(range(1,11)) returns - [1,2,4,5, etc]

List Comprehension - Flattened out for loop

- [num**2 for num in x] returns [1,4,9,16]

Functions - snake_case

- def my_func(param):
    - print(param)

my_func(‘Billy’) - returns - Billy

- or run with default value (param=’default’)


- def new_func(argument)
    - return argument*5
    - gets function out so it can be returned as a variable
- x = new_func(5) - run x - returns 25


Map - Maps a function to an iterable sequence

- seq = [1,2,3,4,5] 
- def times_two(var)
    - return var*2
- list(map(times_two, seq)) reuturns - [2,4,6,8,10]
----------

Lambda Expression - Anonymous function - use one time with no name

- lambda var: var*2

Use Lambda to do the same thing without defining a variable

- list(map(lambda var: var*2, seq))

Filter - Takes in function, applies it to an iterable and returns function items that are true

- def is_even(num):
        return num%2 == 0
    list(filter(is_even, seq)) - returns [2,4]
- Lambda version
- list(filter(lambda num: num%2 == 0, seq)) 

Methods - Call variable.tab to see list of methods

- st = ‘Hello, my name is Billy’
- st.lower() - st.upper()
- tweet = ‘Web3.0 the new internet’
- tweet.split() < splits white space
- tweet.split(‘t’) returns list with ‘t’ remove
```
